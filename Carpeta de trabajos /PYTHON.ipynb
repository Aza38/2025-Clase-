#### 3.- Para el encabezado, extrae empleando expresiones regulares el identificador. (1pto)

```python
# Primero importo la librería "re", que es la que me permite usar expresiones regulares en Python.
# ¿Para qué sirven y qué son? Las expresiones regulares sirven para buscar patrones o fragmentos dentro de un texto.
import re

# Encabezado real (ejemplo del archivo FASTA).
# Mencionado anteriormente, en los archivos FASTA las cabeceras empiezan con el símbolo ">" seguido del identificador del gen o transcrito.
# Ejemplo: >ENST00000625365.2 Contactin associated protein-like 2 (CNTNAP2)
# Lo que quiero es quedarme solo con ENST00000625365.2 (sin el ">").
# ¿Por qué? Porque ese símbolo no forma parte del gen, solo está ahí para indicar que es la cabecera. 
header = ">ENST00000625365.2 Contactin associated protein-like 2 (CNTNAP2)"

# Ahora uso la función re.search() para buscar un patrón dentro del encabezado.
# ^ indica el principio del texto (para empezar desde el inicio)
# > busca el símbolo ">" literal
# (\S+) → los paréntesis indican el grupo que quiero capturar;
#         \S significa "carácter que no sea espacio", y el + indica "uno o más seguidos".
# En resumen: busca el símbolo ">" del principio y captura todo lo que viene después hasta el primer espacio.
# Estoy usando expresiones regulares porque no siempre todos los encabezados son iguales; a veces tienen más espacios o el identificador tiene guiones o puntos.
# Las expresiones regulares son más seguras para reconocer ese patrón sin importar el formato exacto del texto. 
match = re.search(r"^>(\S+)", header)

# .group(1) devuelve el primer grupo capturado entre paréntesis → en este caso, el identificador.
identifier = match.group(1)

# Por último, muestro los resultados en pantalla.
# Primero imprimo el encabezado completo y luego el identificador extraído.
print("Encabezado completo:", header)
print("Identificador extraído:", identifier)
```

#### 4.- ¿Cuál es el porcentaje de GC de la secuencia? (1 pto)

```python
#Me abstengo de explicar otra vez lo que he hecho anteriormente
import re

with open("CNTNAP2.fasta", "r") as f:
    lines = f.readlines()

header = lines[0].replace("\n", "")

match = re.search(r"^>(\S+)", header)
identifier = match.group(1)

sequence = "".join(lines[1:]).replace("\n", "")

length = len(sequence)

#Lo importante ahora: cuento cuántas G y cuántas C hay en la secuencia.
#Este porcentaje mide cuantas GC hay en comparación con el total de la secuencia, que puede contener A y T también. 
#Las G y C se unen con tres puentes de hidrógeno y eso da estabilidad al ADN por eso pide esto el ejercicio. 
g_count = sequence.upper().count("G")
c_count = sequence.upper().count("C")

# Calculo el porcentaje de GC.
#Con estas líneas cuento cuántas G y C hay en toda la secuencia,
#las sumo y calculo qué porcentaje representan sobre el total.
#El resultado me dice cuánto contenido GC tiene el gen,
#que es una medida de estabilidad del ADN y característica de cada gen.

#Lo que hago con g_count + c_count, suma total de G y C. 
#/lenght divide entre la longitud total de la secuencia para saber qué parte representan.
#100 multiplica por 100 para compartirlo en porcentaje. 
gc_content = (g_count + c_count) / length * 100

# Muestro los resultados en pantalla, el header, el identificador, la longitud y el contenido. 
print("Encabezado completo:", header)
print("Identificador extraído:", identifier)
print("Longitud de la secuencia:", length)
print("Porcentaje de GC:", round(gc_content, 2), "%") #Pongo el simbolo de porcentaje para indicar que lo es. 
```

#### 5.- Devuelve la secuencia de ARN que surgiría a partir de la siguiente secuencia. Recuerda que la secuencia de ARN resultante tiene que ser complementaria a la de ADN. (1 pto)

```python
import re
with open("CNTNAP2.fasta", "r") as f:
    lines = f.readlines()

header = lines[0].replace("\n", "")
match = re.search(r"^>(\S+)", header)
identifier = match.group(1)

sequence = "".join(lines[1:]).replace("\n", "")

length = len(sequence)

g_count = sequence.upper().count("G")
c_count = sequence.upper().count("C")
gc_content = (g_count + c_count) / length * 100

print("Encabezado completo:", header)
print("Identificador extraído:", identifier)
print("Longitud de la secuencia:", length)
print("Porcentaje de GC:", round(gc_content, 2), "%")

#Hasta aquí es igual. 

# Ahora quiero obtener la secuencia de ARN complementario.
# El ARN es complementario al ADN, así que sustituyo:
# A → U, T → A, G → C, C → G
conversion_ARN = {"A": "U", "T": "A", "G": "C", "C": "G"}

# Paso toda la secuencia a mayúsculas con el comando upper
sequence_upper = sequence.upper()

# Creo la nueva secuencia de ARN uniendo las bases complementarias.
rna_sequence = "".join([conversion_ARN[base] for base in sequence_upper])

# Muestro los primeros 200 nucleótidos de la secuencia de ARN y su longitud total.
print("\nPrimeros 200 nucleótidos del ARN complementario:")
print(rna_sequence[:200])
print("Longitud total del ARN:", len(rna_sequence))

# Guardo la secuencia de ARN en un nuevo archivo FASTA.
with open("CNTNAP2_RNA.fasta", "w") as f:
    f.write(">CNTNAP2_RNA\n")
    for i in range(0, len(rna_sequence), 100):
        f.write(rna_sequence[i:i+100] + "\n")
```

#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
import re

# Abro el archivo FASTA donde tengo guardada la secuencia del gen CNTNAP2.
# Uso el comando open para abrir el archivo en modo lectura ("r") y readlines() para leer todas las líneas.
with open("CNTNAP2.fasta", "r") as f:
    lineas = f.readlines()

# Me quedo solo con la secuencia de ADN (a partir de la segunda línea, ya que la primera es el encabezado)
# Además, elimino los saltos de línea con replace("\n","").
cadena_adn = "".join(lineas[1:]).replace("\n", "")

# Este diccionario ya me lo han dado.
# Contiene las enzimas de restricción y sus secuencias de reconocimiento.
# El símbolo * marca el punto donde cada enzima corta el ADN.
enzimas_restriccion = {
    'HindIII': 'A*AGCTT',
    'BamHI': 'G*GATCC',
    'AluI': 'AG*CT',
    'Sau3AI': '*GATC',
    'EcoRI': 'G*AATTC'
}

# Defino una función que simula la digestión enzimática del ADN.
# Le paso dos parámetros: la secuencia de ADN (cadena_adn) y la lista de enzimas con sus sitios de corte.
def digestion_adn(cadena_adn, lista_enzimas):
    # Paso toda la secuencia a mayúsculas, por si estuviera mezclada con minúsculas.
    # Esto es importante porque Python distingue entre mayúsculas y minúsculas, y si no, no encontraría bien los sitios de corte.
    cadena_adn = cadena_adn.upper()
    # Creo un diccionario vacío donde se guardará toda la información de los cortes (posiciones, fragmentos, tamaños…).
    info_cortes = {}

    # Recorro el diccionario con las enzimas y sus secuencias de reconocimiento, una por una.
    # Cada elemento tiene un nombre de enzima y su secuencia de reconocimiento.
    for nombre, reconocimiento in lista_enzimas.items():
        # Busco la posición del asterisco * que marca el punto exacto de corte dentro de la secuencia de reconocimiento.
        punto_corte = reconocimiento.find('*')
        # Quito el asterisco porque no forma parte real del ADN.
        secuencia_corte = reconocimiento.replace('*', '')

        # Parte importante: aquí ocurre la búsqueda de los sitios de corte dentro del ADN.
        # Primero, re.finditer() recorre la secuencia completa y encuentra todas las coincidencias del patrón (por ejemplo "GAATTC"),
        # devolviendo la posición donde empieza cada una.
        # Luego, m.start() devuelve esa posición, y le sumo punto_corte, que indica en qué parte exacta del patrón corta la enzima.
        # Todo esto va dentro de una lista por comprensión que guarda todas las posiciones de corte encontradas.
        posiciones_corte = [m.start() + punto_corte for m in re.finditer(secuencia_corte, cadena_adn)]

        # Si hay alguna coincidencia, significa que la enzima sí corta.
        if posiciones_corte:
            # inicio = 0 marca el principio de la secuencia (primer fragmento desde el inicio del ADN).
            # trozos = [] es una lista vacía donde iré guardando los fragmentos de ADN después de cada corte.
            inicio = 0
            trozos = []

            # Este bucle recorre una a una las posiciones donde la enzima corta.
            # Por ejemplo, si corta en [120, 340, 620], el bucle pasará tres veces.
            # En cada vuelta, se guarda el fragmento desde el corte anterior hasta el siguiente.
            for pos in posiciones_corte:
                trozos.append(cadena_adn[inicio:pos])
                inicio = pos

            # Al terminar, añado el último fragmento (desde el último corte hasta el final del ADN).
            trozos.append(cadena_adn[inicio:])

            # Guardo los resultados en el diccionario info_cortes para esa enzima.
            # Si la enzima corta, guardo True, las posiciones exactas, los fragmentos obtenidos y sus longitudes.
            info_cortes[nombre] = {
                'corta': True,
                'posiciones': posiciones_corte,
                'fragmentos': trozos,
                'longitudes': [len(t) for t in trozos]
            }
        else:
            # Si no corta, también lo guardo, pero indicando que no hay cortes.
            # En ese caso, la lista de posiciones está vacía y la secuencia completa queda como un solo fragmento.
            info_cortes[nombre] = {
                'corta': False,
                'posiciones': [],
                'fragmentos': [cadena_adn],
                'longitudes': [len(cadena_adn)]
            }

    # Devuelvo toda la información con los resultados de la digestión para cada enzima.
    return info_cortes

# Ejecuto la digestión sobre mi secuencia y mis enzimas.
# El resultado se guarda en resultado_digestion.
resultado_digestion = digestion_adn(cadena_adn, enzimas_restriccion)

# Muestro los resultados por pantalla.
# Este bucle recorre una por una las enzimas y muestra toda la información guardada.
for nombre, datos in resultado_digestion.items():
    # Muestra el nombre de la enzima y si corta (True) o no (False)
    print(f"\nEnzima: {nombre}")
    print("¿Corta la secuencia?", datos['corta'])
    # Si la enzima sí corta, muestro información más detallada.
    if datos['corta']:
        # Posiciones exactas de corte, número de fragmentos, tamaños de cada uno y los primeros nucleótidos del primer fragmento.
        print("Posiciones de corte:", datos['posiciones'])
        print("Número de fragmentos:", len(datos['fragmentos']))
        print("Tamaños de fragmentos:", datos['longitudes'])
        print("Primeros 200 nucleótidos del primer fragmento:", datos['fragmentos'][0][:200])
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python

with open("CNTNAP2.fasta", "r") as f:
    lineas = f.readlines()
secuencia_adn = "".join(lineas[1:]).replace("\n", "").upper()

# Defino los codones de inicio, el de inicio es ATG además que siempre es así y parada, que son siempre cualquiera de esos codones. 
inicio = "ATG"
paradas = ["TAA", "TAG", "TGA"]

# Esta es la función para buscar ORFs, que recorre la seucnecia del ADN buscando los marcos posibles de lectura a los que llamamos ORFs
#Son las regiones del ADN que podrían codificar una proteína, porque están entre un codón de inicio y uno de parada. 
def buscar_orfs(adn):
    orfs = [] #Creo una lista vacía donde guardo todos los ORFs que encuentre el programa.
    for i in range(len(adn)): #Recorro la secuencia base por base. Este bucle recorre la secuencia de ADN posición por posición. 
        #range(len(adn)) crea una secuencia desde 0 hasta la longitud del ADN menos uno. Es decir i toma los valores 0, 1, 2, 3, ...así hasta el final. 
        #Eso significa que estoy avanzando base por base a lo largo de la molécula del adn como si fuera leyendo nucleótido a nucleótido hasta encontrar el inicio de un gen. 
        if adn[i:i+3] == inicio: #significa “toma un trozo de tres letras desde la posición i”.
#Es decir, estoy extrayendo un codón, que son grupos de tres bases.
#Luego comparo si ese triplete es igual a "ATG", que está guardado en la variable inicio.
#Lo que estoy haciendo es que ATG es el codón de inicio que es donde se marca que empieza un gen entonces si encuentro ese codón significa que puede empezar ahí un ORF. 
            for j in range(i+3, len(adn), 3): #Este segundo bucle empieza tres posiciones después del inicio i, osea i + 3 y avanza de tres en tres, step = 3 porque siempre tienen tres bases. 
                #Eso es que estoy leyendo a partir del codón de inicio como si fuese un ribosoma codón por codón en grupos de tres bases buscando el codón de parada que indica el final de la proteína. 
                if adn[j:j+3] in paradas: #Aquí estoy tomando un grupo de tres bases (adn[j:j+3])
#y compruebo si ese triplete está en la lista paradas = ["TAA", "TAG", "TGA"].
#Si lo está, significa que he llegado a un codón de parada. 
#Es el momento en que termina el gen: el ribosoma dejaría de traducir la proteína.
#He encontrado el final del ORF, y puedo guardar ese fragmento completo del ADN.           
                    orf = adn[i:j+3] #Creo una nueva variable llamada orf que contiene la porción de ADN
#desde el inicio del gen (i) hasta el final (j+3, que incluye el codón de parada).
                    #Este fragmento de ADN representa una posible región codificante completa:
#desde el codón de inicio (ATG) hasta el de parada (TAA, TAG o TGA).
#Es decir, un marco abierto de lectura (ORF).
                    orfs.append(orf) #.append() añade ese ORF a la lista orfs, que contiene todos los encontrados.
#Voy guardando todas las posibles regiones que podrían corresponder a genes,
#ya que un mismo ADN puede tener varios ORFs.
                    break #El comando break hace que salga del bucle interno (el que avanza de tres en tres).
#De esta forma, una vez encontrado el primer codón de parada después de un inicio,
#no sigo buscando más paradas dentro del mismo ORF
#En un gen real, cuando se alcanza un codón de parada, la traducción termina.
#Por eso no sigo más allá: ese ORF ya está completo.

#HASTA AQUÍ EXPLICACIÓN: ste bloque de código simula cómo una célula leería el ADN:
#busca un inicio (ATG).
#Empieza a leer codones (tripletes) de tres en tres.
#Cuando encuentra un codón de parada, guarda ese fragmento como un ORF.
#Continúa buscando nuevos posibles genes en la secuencia.
    return orfs

# Encuentro los ORFs, ejecuta la función que he definido antes para buscar. La devuelve en una lista con todos los posibles ORFs. 
#Ya tengo todas las regiones codificantes de mi gen CNTNAP2
#Cada región puede ser un gen completo o una parte de él dependiendo de donde empiece y donde acabe. 
orfs = buscar_orfs(secuencia_adn)

# Ordeno los ORFs por longitud (de mayor a menor) Con sorted ordena según longitud key =len
#Reverse=true es para que empiece por el más largo
#Los genes más largos suelen ser probablemente los más funcionales, porque los ORFs cortos muchas veces son solo fragmentos que no significan nada
#o falsos positivos. Y así priorizas los de mayor probabilidad.
orfs_ordenados = sorted(orfs, key=len, reverse=True)

# Va a coger los 6 más largos los seis primeros de la lista. 
#Me quedo con las 6 regiones más grandes, las que interesan. 
orfs_top6 = orfs_ordenados[:6]

#El comando de enumerate es para recorrer los seis ORFs y ponerles número empezando desde 1 
#(porque estoy usando números normales, fuera de programación) Y luego lo imprime junto con su longitud en bases. 
#Estoy comparando el tamaño de los seis posibles genes que he encontrado. Esto debe coincidir con un estudio real si miro en alguno de esos ORFs que coincida con un gen conocido en bases de datos. 
print("Los 6 ORFs más largos:")
for i, orf in enumerate(orfs_top6, 1):
    print(f"{i}. Longitud: {len(orf)} bases")

# Me quedo con el más largo como mejor predicción. Eligo el mejor de todos, el candidato con más probabilidades.
mejor_orf = orfs_top6[0]

# Ahora hago la traducción a proteína 
# Diccionario mínimo para traducir ADN a aminoácidos donde cada triplete de ADN (codón) se asocia con el aminoácido que codifica. 
tabla_codon = {
    "ATG":"M", "TTT":"F", "TTC":"F", "TGG":"W", "TAT":"Y", "TAC":"Y"
}
# Solo incluyo los aromáticos y el inicio para simplificar. ATG por ejemplo es M 
#Cada tres bases del ADN forman un codón, que el ribosoma traduce a un aminoácido durante la síntesis de proteínas.
#En este caso solo usamos los aminoácidos aromáticos (Fenilalanina F, Triptófano W y Tirosina Y) y la metionina (M),
#porque son importantes y fáciles de identificar.
#Los aromáticos tienen anillos en su estructura química, y suelen participar en la estabilidad y plegamiento de las proteínas.

#creo la función, la llamo así. Empiezo con una cadena vacía donde guardaré los aminoácidos.
def traducir_adn(adn):
    proteina = ""
    #Recorre la secuencia de 3 en tres porque cada codon tiene esas bases. 
    for i in range(0, len(adn)-2, 3):
        #Esto lo he explicado antes, coge un trozo de 3 bases. 
        codon = adn[i:i+3]
        #Busca ese codón en el diccionario y si existe le añade el aminoácido que correspone.
        #Y si no, pone un guión. Pongo eso porque ya no se me ocurre qué poner, aunque no se si es correcto usarlo porque igual confunde. 
        #Devuelve toda la secuencia de aminoácidos en la variable proteína. 
        proteina += tabla_codon.get(codon, "-")  # "-" si no está en la tabla
    return proteina
#Esta función simula lo que haría el ribosoma en una célula:
#va “leyendo” el ARN (que aquí representamos con la secuencia de ADN)
#de tres en tres bases (codones), y por cada codón añade el aminoácido correspondiente a la proteína.
#Cuando termina de leer, ya tiene una cadena de aminoácidos, que es la proteína final.
proteina = traducir_adn(mejor_orf)
#Ya he conseguido traducir mi mejor predicción de gen (el ORF más largo)
#a su correspondiente secuencia de aminoácidos,
#es decir, la posible proteína que codifica ese gen.

# Los cálculos:
# Porcentaje de aromáticos (F, W, Y)
#count para contar los aromáticos y su porcentaje. 
aromaticos = proteina.count("F") + proteina.count("W") + proteina.count("Y")
porcentaje_aromaticos = (aromaticos / len(proteina)) * 100

# Longitud relativa (cuánto ocupa el ORF en el ADN)
porcentaje_longitud = (len(mejor_orf) / len(secuencia_adn)) * 100
#El contenido aromático ayuda a entender cómo es la proteína:
#los aminoácidos con anillos (F, W, Y) son importantes para la estructura interna,
#porque estabilizan el plegamiento por interacciones hidrofóbicas y apilamiento de anillos.
#Una proteína con muchos aromáticos suele tener una estructura más rígida o estable.
#Indica qué porcentaje del gen analizado corresponde a una región codificante real.
#Por ejemplo, si da un 15%, quiere decir que el posible gen ocupa el 15% de toda la secuencia del ADN del archivo.
    
#print de los resultados. 
print("\nMejor ORF encontrado:")
print("Longitud:", len(mejor_orf), "bases")
print("Primeros 100 aminoácidos:", proteina[:100])
print("Porcentaje de aminoácidos aromáticos:", round(porcentaje_aromaticos, 2), "%")
print("Longitud relativa respecto al ADN:", round(porcentaje_longitud, 2), "%")
#Estos resultados resumen el análisis funcional del gen CNTNAP2:
#El ORF más largo representa la predicción más probable del gen real.
#La secuencia de aminoácidos es su posible proteína funcional.
#El porcentaje de aromáticos da una pista sobre su estructura tridimensional.
#Y la longitud relativa indica qué parte del ADN realmente codifica información útil.
    
```
